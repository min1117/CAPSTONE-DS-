```python
cov_mat = np.cov(df2_std .T)
eigen_vals,eigen_vecs = np.linalg.eig(cov_mat)
print('/n 고유값 /n%s' % eigen_vals)
```

```python
tot =sum(eigen_vals)
var_exp = [(i / tot ) for i in
          sorted(eigen_vals,reverse=True)]
cum_var_exp = np.cumsum(var_exp)

plt.bar(range(1,6),var_exp,alpha=0.5,align='center',label='individual explained variance')
plt.step(range(1,6),cum_var_exp,alpha=0.5,where='mid',label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
distortions = []
for i in range(1,11):
    km = KMeans(n_clusters = i,
               init = 'k-means++',
               n_init = 10,
               max_iter = 300,
               random_state = 0)
    km.fit(scores_pca)
    distortions.append(km.inertia_)
plt.plot(range(1,11),distortions,marker = 'o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
km = KMeans(n_clusters=4, init='k-means++',n_init = 10, max_iter=300,random_state=0)
y_km = km.fit_predict(scores_pca1)

from sklearn.metrics import silhouette_samples
from matplotlib import cm
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X,
                                    y_km,
                                    metric = 'euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper),
            c_silhouette_vals,
            height=1.0,
            edgecolor='none',
            color=color)
    yticks.append((y_ax_lower + y_ax_upper) /2.)
    y_ax_lower += len(c_silhouette_vals)
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg,
           color = 'red',
            linestyle="--")
plt.yticks(yticks,cluster_labels +1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.tight_layout()
plt.show()
```

df3의 특성이 6개이므로 6x6차원의 공분산 행렬로 6개의 고유벡터와 고유값을 얻을 수 있다.

np.cov함수를 사용하여 표준화 전처리된 데이터셋의 공분산 행렬을 계산한다. 
linalg.eig함수를 사용하여 고유값 분해를 수행하고 이를 통해 6개의 고유값이 들어있는 벡터와 각 고유값에 대응하는 고유벡터가 열에 저장된 
6x6차원의 행렬을 얻었다. 

```python
tot =sum(eigen_vals)
var_exp = [(i / tot ) for i in
          sorted(eigen_vals,reverse=True)]
cum_var_exp = np.cumsum(var_exp)

plt.bar(range(1,7),var_exp,alpha=0.5,align='center',label='individual explained variance')
plt.step(range(1,7),cum_var_exp,alpha=0.5,where='mid',label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

데이터셋 차원을 새로운 특성 부분 공간으로 압축하여 줄여야 하기 때문에 가장 많은 정보(분산)을 가진 고유 벡터(주성분)일부만을 선택한다.
고윳값은 고유 벡터의 크기를 결정하므로 고윳값을 내림차순으로 정렬한다. 고윳값 순서에 따라 최상위 k개의 고유 벡터를 선택한다. 가장 정보가 많은 k개의 고유 벡터를 선택하기 전에 고윳값이 설명된 분산비율을 그래프로 그린것이다.(395) 
결과 그래프는 첫 번째 주성분이 거의 분산의 60%를 설명하고 있고 처음 두 개(1,2)의 주성분이 데이터셋에 있는 분산의 대략 80%를 설명한다.

```python
from sklearn import metrics
from sklearn.cluster import KMeans

n_clusters = [2,3,4,5,6,7,8,9]
for k in n_clusters:
    kmeans = KMeans(n_clusters = k, random_state = 0).fit(scores_pca2)
    cluster_labels = kmeans.predict(scores_pca2)
    S = metrics.silhouette_score(scores_pca2, cluster_labels)
    print("n_clusters = {:d}, silhouette score {:1f}".format(k,S))
```
